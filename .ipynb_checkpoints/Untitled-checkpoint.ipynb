```python
# Importation des bibliothques ncessaires
from sklearn.datasets import make_moons 
# Module pour gnrer des donnes en forme de lune
import matplotlib.pyplot as plt 
# Bibliothque pour la cration de graphiques
from sklearn.preprocessing import StandardScaler 
# Module pour la standardisation des donnes

# Gnration de donnes 
X, _ = make_moons(n_samples=400, noise=0.1, random_state=42)

X = StandardScaler().fit_transform(X)

plt.figure(figsize=(8, 6))

plt.scatter(X[:, 0], X[:, 1])  
plt.title(' Dataset for DBSCAN')
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2')  
plt.grid(True) 
# affichage 
plt.show()
```
